```python
def state(x):
  st = x.split('(')[1]
  return st[:-1]
```

```python
def city(x):
  ct = x.split('(')[0]
  ct = ct.split('_')[0]
  return ct
```

```python
sort_segment['destination_city'] = sort_segment['destination_name'].apply(lambda x: city(x))
```

```python
def place(x):
  x = x.split('(')[0]
  len_ = len(x.split("_"))

  if len_ >= 3:
    return x.split('_')[1]

  if len_ == 2:
    return x.split('_')[0]

  return x.split(" ")[0]
```

```python
sort_segment['destination_place'] = sort_segment['destination_name'].apply(lambda x: place(x))
```

```python
sort_segment['destination_place']
```

```python
def code(x):
  x = x.split('(')[0]

  if len(x.split("_")) >=3:
    return x.split('_')[-1]

  return 'none'
```

```python
for col in numerical_features:
  Q1 = trip_data[col].quantile(0.25)
  Q3 = trip_data[col].quantile(0.75)
  IQR = Q3 - Q1
  LW = Q1 - 1.5 * IQR
  UW = Q3 + 1.5 * IQR
  trip_data = trip_data[(trip_data[col]>=LW) & (trip_data[col]<=UW)]
```

```python
ot_mean = trip_data['osrm_time']
ot_mean
```

Null Hypothesis: Actual time and OSRM time are statistically similiar to each other.
Alternate Hypothesis: Actual time and OSRM time are statistically different from each other.

```python
from scipy.stats import ttest_ind
```

```python
ttest_ind(at_mean,ot_mean)
```

Since, P-value is greater than alpha value i.e 0.05.
Hence, Null Hypothesis is true.
So, Actual time and OSRM time values are similiar.
